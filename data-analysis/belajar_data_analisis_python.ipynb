## Tipe Data Boolean dan Operator Perbandingan
Boolean (atau bool) adalah tipe data yang hanya memiliki dua nilai, yaitu True atau False. Biasanya digunakan untuk menyatakan kondisi logika atau hasil operator perbandingan. Operator perbandingan digunakan untuk membandingkan dua nilai, dan hasilnya selalu berupa boolean `True` atau `False`.

```python
4 > 2 # lebih besar dari
4 < 2 # lebih kecil dari
4 == 4 # sama dengan
3 >= 3 # lebih besar atau sama dengan
4 <= 5 # lebih kecil atau sama dengan
4 != 5 # tidak samadengan
```

## Tipe Data Tuple
Tuple adalah tipe data yang digunakan untuk menyimpan sekumpulan nilai seperti List, tetapi bersifat immutable (tidak dapat diubah setelah dibuat). Nilainya ditulis di dalam tanda `(elemen1, elemen2, elemen3)`.

**Kesimpulan:** Tuple ibaratkan sebuah wadah data seperti list, tapi isinya tetap (tidak bisa diubah). Cocok digunakan untuk data yang bersifat tetap atau tidak boleh dimodifikasi.

## Tipe Data Set
Set adalah tipe data yang digunakan untuk menyimpan kumpulan nilai unik (tidak ada duplikat) dan tidak memiliki urutan indeks. Nilainya ditulis dalam `{elemen1, elemen2, elemen3}`.

```python
s = {1,1,1,2,2,3,'a','a','b'}
print(s)
```

Kita juga bisa menggunakan operasi himpunan seperti Union, Intersection, dan Difference.

```python
A = {1, 2, 3}
B = {3, 4, 5}

print(A | B)  # Union → {1, 2, 3, 4, 5}
print(A & B)  # Intersection → {3}
print(A - B)  # Difference → {1, 2}
```

**Kesimpulannya:** Set adalah sekumpulan elemen unik tanpa urutan, cocok untuk operasi matematika himpunan atau saat kamu ingin menghilangkan data yang duplikat.

## Pengulangan (Loop Statement)
Pengulangan atau Loop Statement ini digunakan untuk mengulang blok kode hingga setiap elemennya habis atau tercapai kondisi tertentu. Loop Statement ini ada 2 yang bisa digunakan, yang pertama adalah **for loop** dan **while loop**.

```python
# for loop: mengulang blok kode dengan objek sequential, contoh list, string, taple, dan lainnya

var =  [1,2,3,4,5]

for item in var:
  #blok kode
  print(item)
  print("halo")
  print(item**3)
```

```python
# while loop: mengulang blok kode sampai kondisi tertentu tercapai

i = 4

while i > 0:
  #blok kode
  print('halo')
  i = i - 1
  print(i)
```

```python
# loop control statement: break

i = 4

while True:
  #blok kode
  print('halo')
  i = i - 1
  print(i)
  if i == 0:
    break
```

## List Comprehension

## Distribusi Probabilitas (Penjelasan Intuitive)
Distribusi Probabilitas digunakan untuk menggambarkan bagaimana nilai dari suatu variabel acak didistribusikan. Variabel acak adalah suatu nilai yang dihasilkan dari eksperimen acak, seperti lemparan dadu, hasil ujian, atau tinggi badan dalam populasi tertentu.

Distribusi Uniform

Distribusi Uniform adalah distribusi dimana semua hasil yang mungkin dari suatu variabel acak memiliki probabilitas yang sama.

Distribusi Normal

Distribusi Normal, juga dikenal sebagai distribusi Gaussian, adalah salah satu distribusi paling penting dalam statistik. Distribusi ini berbentuk seperti lonceng (bell curve) dan simetris terhadap mean (rata-rata).

Distribusi Binomial

Distribusi Biomial adalah distribusi diskrit yang menggambarkan jumlah sukses dalam sejumlah percobaan independen yang hanya memiliki dua hasil (misalnya sukses atau gagal).

Parameter pada distribusi normal:

 `n = jumlah percobaan` dan `p = probabilitas` sukses dalam satu percobaan.

Distribusi Poisson

Distribusi Poisson adalah distribusi diskrit yang digunakan untuk memodelkan jumlah kejadian dalam interval waktu atau ruang tertentu ketika kejadian-kejadian tersebut terjadi secara independen dengan rata-rata konstan.

Parameter pada distribusi poisson:

`λ (lamda) = rata-rata jumlah kejadian dalam satu interval`

## Cara Menggabungkan DataFrame (Concat, Marge dan Join)
Terdapat tiga cara untuk menggabungkan DataFrame di library Pandas, yaitu menggunakan fungsi Concat, Merge, dan Join).

```python
# Membuat tiga DataFrame

df1 = pd.DataFrame({
    'key':['A','B','C'],
    'value1':[1,2,3]
})

df2 = pd.DataFrame({
    'key':['D','E','F'],
    'value1':[4,5,6]
})

df3 = pd.DataFrame({
    'value2':[4,5,6]
})
```

```python
pd.concat([df1,df3],axis=1)
```

```python
dfA = pd.DataFrame({
    'key':['A','B','C','D'],
    'value1':[1,2,3,4]
})

dfB = pd.DataFrame({
    'key':['A','B','E','F'],
    'value2':[5,6,7,8]
})

dfC = pd.DataFrame({
    'value1':[1,2,3,4]
}, index=['A','B','C','D'])

dfD = pd.DataFrame({
    'value2':[5,6,7,8]
}, index=['A','B','E','F'])
```

```python
# Fungsi Merge
pd.merge(dfA, dfB, on='key', how='left')
```

```python
# Fungsi Join
dfC.join(dfD, how='left')
```

## Plot-plot Lainnya
Plot lainnya yang tersedia di matplotlib dapat dilihat di:

Cari diagram yang mau kamu pake, copy, paste disini

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

n_bins = 10
x = np.random.randn(1000, 3)

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

colors = ['red', 'tan', 'lime']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')

ax1.hist(x, n_bins, density=True, histtype='bar', stacked=True)
ax1.set_title('stacked bar')

ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('stack step (unfilled)')

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('different sample sizes')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns = None

plt.style.use('ggplot')

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sample_data/dummy_ojol_transactions.csv', sep=';')
df.head(1)
```

Dara data diatas ditemukan tipe data yang tidak sesuai, yaitu "distance" dimana ada data "5,55". Kita tahu bahwa seharusnya aturan yang benar di python itu "5.55". Jadi kita bisa asumsikan bahwa ini adalah tipe object string, untuk memastikannya gunakan fungsi info().

```python
df.info()
```

```python
df['distance'].astype('float') # kita coba mengubahnya menjadi float, tetapi error karena ada data yang menggunakan koma (,)
```

Untuk mengatasinya kita mengubah data yang ada koma-nya menjadi titik

```python
df['distance'] = df['distance'].str.replace(',','.')
df['distance'] = df['distance'].astype('float') # nah hasilnya berhasil
```

```python
fig, ax = plt.subplots()

df[df['mode']=='BIKE'].plot(kind='scatter',x='distance',y='transaction_amount_total',color='g',label='BIKE', ax=ax)
df[df['mode']=='CAR'].plot(kind='scatter',x='distance',y='transaction_amount_total',color='r',label='CAR', ax=ax)

fig.show()
```

# Plotly: Modifikasi dan Macam-macam Plot Plotly

## Modifikasi Layout Part 1

Method yang digunakan untuk melakukan modifikasi layout pada plotly express adalah `update_layout()`

```python
import plotly.express as px

iris = px.data.iris()
iris.head(3)
```

```python
fig = px.scatter(iris,
           x='sepal_width',
           y='sepal_length',
           color='species',
           title="Scatter Plot"
           )

fig.update_layout(
    title = {
        'text' : 'Scatter Plot',
        'y' : 0.87,
        'x' : 0.5,
        'xanchor' : 'center',
        'yanchor' : 'top'
    },
    xaxis = {
        'title' : 'Sepal Width'
    },
    yaxis = {
        'title' : 'Sepal Length'
    }
)
```

```python
fig = px.scatter(iris,
           x='sepal_width',
           y='sepal_length',
           color='species',
           title="Scatter Plot"
           )

title_dic = {
        'text' : 'Scatter Plot',
        'subtitle': {
            'text':'Scatter Plot Iris Subtitle',
            'font': {
                'size': 14
            }
            },
        'y' : 0.87,
        'x' : 0.5,
        'xanchor' : 'center',
        'yanchor' : 'top'
    }

xaxis_dic = {
        'title' : 'Sepal Width'
    }

yaxis_dic = {
        'title' : 'Sepal Length'
}

fig.update_layout(
    title = title_dic,
    xaxis = xaxis_dic,
    yaxis = yaxis_dic
)
```

## Modifikasi Layout Part 2

```python
fig = px.scatter(iris,
           x='sepal_width',
           y='sepal_length',
           color='species'
           )

fig.update_layout(
    title_text="Scatter Plot",
    title_subtitle_text='Iris Dataset',
    title_y = 0.97,
    title_x = 0.5,
    title_xanchor = 'center',
    xaxis_title = 'Sepal Width',
    yaxis_title = 'Sepal Length',
    legend_title = 'Species',
    title_font_color = 'red'
)

fig.show()
```

## Modifikasi Traces
Biasa digunakan untuk memodifikasi hal-hal yang berhubungan sama plotnya.

link: https://plotly.com/python/reference/index/

```python
import plotly.express as px

df = px.data.gapminder().query("year==2007 and continent=='Asia'")
df.head()
```

- country: Nama negara, seperti "Indonesia", "India", atau "Brazil". Ini adalah kolom kategorikal yang menunjukkan negara yang sedang dianalisis.

- continent: Benua tempat negara tersebut berada, seperti "Asia", "Europe", "Africa", "Americas", atau "Oceania".

- year: Tahun dalam bentuk numerik, yang biasanya mencakup rentang dari sekitar tahun 1950-an hingga 2000-an. Ini adalah kolom waktu yang menunjukkan tahun pengukuran data.

- lifeExp (Life Expectancy): Harapan hidup rata-rata di negara tersebut pada tahun tertentu, diukur dalam tahun.

- gdpPercap (GDP per Capita): Produk domestik bruto (PDB) per kapita, atau pendapatan rata-rata per orang dalam negara tersebut dalam satu tahun. Biasanya dinyatakan dalam dolar AS.

- pop (Population): Populasi total negara pada tahun tersebut, dinyatakan dalam jumlah orang.

Mengubah Warna Marker

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', color='continent', title="GDP vs Life Expectancy")
fig.update_traces(marker_color='red')  # Mengubah warna marker menjadi merah
fig.show()
```

Mengubah Ukuran Marker

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent', title="GDP vs Life Expectancy with Population Size")
fig.update_traces(marker_size=20)  # Mengubah ukuran marker menjadi 20
fig.show()
```

Mengubah Bentuk Marker

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', color='continent', title="GDP vs Life Expectancy")
fig.update_traces(marker_symbol='diamond')  # Mengubah bentuk marker menjadi diamond
fig.show()
```

Transparansi

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', color='continent', title="GDP vs Life Expectancy")
fig.update_traces(opacity=0.5)  # Mengatur transparansi marker menjadi 0.5
fig.show()
```

Mengubah Posisi Text

```python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", text="country", log_x=True, size_max=60)

fig.update_traces(textposition='top center')

fig.update_layout(
    height=800,
    title_text='GDP and Life Expectancy (Americas, 2007)'
)

fig.show()
```
